1. What is Simple Linear Regression (SLR)? Explain its purpose.
   - Simple Linear Regression (SLR) is a statistical technique that models the relationship between one independent (predictor) variable and one dependent (response) variable using a straight line.
   Main Purpose
-To understand the linear relationship between two quantitative variables, especially the direction (positive/negative) and strength of association via the slope.
-To predict the expected value of the dependent variable for a given value of the independent variable by using the fitted regression line.



2. What are the key assumptions of Simple Linear Regression?
   - Simple Linear Regression (SLR) rests on a small set of standard assumptions about the form of the relationship and the behavior of errors. These ensure that estimates, confidence intervals, and hypothesis tests are valid.
   Core model assumptions
Linearity: The true relationship between the predictor X and the mean of the response Y is a straight line, i.e., E(Y/X)=β0+β1X.

Additive, zero-mean errors: The model is Y=β0+β1X+ε where the error term ε has expected value 0 for every value of X.


3. Write the mathematical equation for a simple linear regression model and
explain each term.
   - A simple linear regression model is usually written as
   Yi=β0+β1Xi+εi for each observation i.
Yi: The dependent (response) variable for the i-th observation; the outcome being modeled or predicted.

Xi: The independent (predictor/explanatory) variable for the i-th observation; used to explain or predict Yi.

β0: The intercept, meaning the expected value of Y when X=0; it is a fixed but unknown model parameter.

β1: The slope coefficient; it represents the average change in Y for a one-unit increase in X.

εi: The random error term for observation i, capturing all other unmodeled influences and randomness; it is assumed to have mean 0 and some variance σ2.



4. Provide a real-world example where simple linear regression can be
applied.
   - A classic real-world use of simple linear regression is predicting house price from house size (area in square feet or square meters).

Scenario
A real estate analyst has past sales data: for each house, they know its size (e.g., in square feet) and its selling price.

Plotting price vs size usually shows that larger houses tend to sell for more, roughly following a straight-line trend.
How SLR is applied
Treat house size as the independent variable X and house price as the dependent variable Y.

Fit a simple linear regression model Y=β0+β1X+ε to estimate how much the average price changes for each additional unit of size and to predict the price of a new house given its size.



5.What is the method of least squares in linear regression?
   - The method of least squares is the standard way to estimate the line of best fit in linear regression by minimizing the sum of squared prediction errors (residuals).



6. What is Logistic Regression? How does it differ from Linear Regression?
   - Logistic Regression is a statistical and machine learning method used to model the probability that a dependent variable belongs to a particular class, most commonly for binary outcomes (0/1, yes/no, churn/no churn). It outputs values between 0 and 1 that are interpreted as probabilities and then converted to classes via a threshold (often 0.5).
   In practice:

-Use Linear Regression when the target is numeric and unbounded (e.g., predicting house price).

-Use Logistic Regression when the target is categorical, especially binary (e.g., predicting whether a customer will churn: yes/no).



7. Name and briefly describe three common evaluation metrics for regression
models.
   - Three common evaluation metrics for regression models are Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and R-squared (R²). These metrics assess prediction accuracy and model fit by comparing actual and predicted values.
   Mean Absolute Error (MAE)
MAE calculates the average absolute difference between predicted and actual values, given by
MAE=1n∑i=1n|yi-y^i|.
Root Mean Squared Error (RMSE)
RMSE is the square root of the average squared differences,
It penalizes larger errors more heavily due to squaring and is widely used as it matches the scale of the data.
R-squared (R²)
R², or coefficient of determination, measures the proportion of variance in the dependent variable explained by the model, ranging from 0 to 1 (higher is better).



8. What is the purpose of the R-squared metric in regression analysis?
   - R-squared (R²), also known as the coefficient of determination, measures the proportion of the total variance in the dependent variable that is explained by the independent variables in a regression model.

Primary Purpose
R² quantifies model fit by comparing the model's predictions to simply using the mean of the observed values, with values ranging from 0 (no explanatory power) to 1 (perfect fit). Higher values indicate the model captures more variability in the data, helping assess goodness-of-fit.






```python
#9.Write Python code to fit a simple linear regression model using scikit-learn
#and print the slope and intercept.
#(Include your Python code and output in the code box below.)
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data: house size (sq ft) vs price (thousands $)
X = np.array([[1400], [1600], [1700], [1875], [2025], [2150],
              [2400], [2825], [2900], [3000]]).reshape(-1, 1)
y = np.array([245, 312, 279, 308, 199, 219, 405, 324, 319, 342])

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Print slope (beta1) and intercept (beta0)
print(f"Slope (β₁): {model.coef_[0]:.4f}")
print(f"Intercept (β₀): {model.intercept_:.4f}")

# Optional: Visualize the fit
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (thousands $)')
plt.legend()
plt.show()
```

10.  How do you interpret the coefficients in a simple linear regression model?
   - In simple linear regression, the coefficients—the intercept (β0) and slope(β1)—describe the fitted line Y^=β0+β1X
Intercept (β0)
The intercept represents the predicted value of the dependent variable Y when the independent variable X=0. Its practical meaning depends on whether X=0 is realistic in context (e.g., meaningless for house prices if size cannot be zero).
Slope (β1)
The slope indicates the average change in Y for each one-unit increase in X. A positive β1 shows a direct relationship (as X rises, Y rises); negative shows inverse. For example, if β1=0.3 in a house price model, each additional square foot increases predicted price by $300 (in thousands).
Key Interpretation Rules
Coefficients assume other variables (none in SLR) are held constant.

Statistical significance (via p-value < 0.05) confirms the relationship is unlikely due to chance.

Always check units: if X is in square feet and Y in thousands of dollars, β1 is dollars per square foot (scaled).
